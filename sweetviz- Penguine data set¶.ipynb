# Penguine data set 

## 1. Read in the Penguins Dataset

```python
import pandas as pd
penguins_df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/penguins_cleaned.csv')
```

```python
# Separating X and y
X = penguins_df.drop('species', axis=1)
y = penguins_df['species']
```

## 2. Analyze the Penguins Dataset

```python
penguins_df
```

```python
import sweetviz as sv
analyze_report = sv.analyze(penguins_df)
analyze_report.show_html('analyze.html', open_browser=False)
```

```python
import IPython
IPython.display.HTML('analyze.html')
```

## 3. Side-by-Side Comparison of Train vs Test Set

### 3.1. Data Split using 80/20 Split Ratio

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_test
```

### 3.2. Comparison Report

```python
import sweetviz as sv
compare_report = sv.compare([X_train, 'Train'], [X_test, 'Test'])
compare_report.show_html('compare.html', open_browser=False)
```
